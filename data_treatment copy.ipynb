```python
# Import the necessary library
import json

# Read the dictionary from the text file
with open('topic_dictionary.txt', 'r') as file:
    topic_hashtags = json.load(file)
```

```python
for topic in topic_hashtags.keys():
    print('-----------',topic,"-----------")
    for hashtag in topic_hashtags[topic].keys():
        
        list_dict = extract_mastodon(mastodon, hashtag,50 )
        topic_hashtags[topic][hashtag]=list_dict
        print('                         ',hashtag, len(list_dict))
```

```python
import pandas as pd

# Your nested dictionary
# author +  create_at + contents + hashtag_related
# Initialize an empty list to store records
records = []

# Flatten the nested dictionary into records
for topic, hashtags in topic_hashtags.items():
    for hashtag, entries in hashtags.items():
        for entry in entries:
            records.append({
                "topic": topic,
                "hashtag": hashtag,
                "content": entry["content"],
                "hashtag_related 1": entry["related_topics"][0],
                "hashtag_related 2": entry["related_topics"][1],
                "hashtag_related 3": entry["related_topics"][2],
                "hashtag_related 4": entry["related_topics"][3],
                "hashtag_related 5": entry["related_topics"][4],
                "author": entry["author"],
                "create_at": entry["created_at"],
            })

# Create a Pandas DataFrame
df = pd.DataFrame(records)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')

tokenized_contents = [word_tokenize(content) for content in contents]
```

```python
tokenized_contents[0]
```

### part of speach tagging

```python
nltk.download('averaged_perceptron_tagger')
pos_tagged = [nltk.pos_tag(sentence) for sentence in tokenized_contents]
print(pos_tagged)
```

### Named Entity Recognition (NER): You can use NLTK to identify named entities (people, places, organizations, etc.) in your text.

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
ner_sentences = [nltk.ne_chunk(sentence) for sentence in pos_tagged]
print(ner_sentences)
```

### Stemming and Lemmatization

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_sentences = [[stemmer.stem(word) for word in sentence] for sentence in tokenized_contents]
print(stemmed_sentences)
```

### Building a Bag-of-Words Models

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform([" ".join(sentence) for sentence in tokenized_contents])
print(X.toarray())
```

### Building Word Embeddings: You can train a Word2Vec model on your sentences to create word embeddings that capture semantic meanings of words.

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences=tokenized_contents, vector_size=100, window=5, min_count=1, workers=4)
print(model.wv["math"])
```
